#### Example: generating random samples from the Normal distribution.

```python
import numpy as np
import matplotlib.pyplot as plt
from reliapy.distributions.continuous import Normal
from reliapy.distributions.joint_distribution import JointDistribution
```

```python
# Marginal distributions.
M0 = Normal(loc=40, scale=5)
M1 = Normal(loc=50, scale=2.5)

# Correlation matrix.
Cz = np.array([[1.0, 0.6], [0.6, 1.0]])

# Instantiate an object of JointDistribution.
JD = JointDistribution(marginal=[M0, M1], correlation=Cz, random_state=1111)

# Get random samples.
x = JD.rvs(n_sim=10000)
```

```python
# Plot
plt.scatter(x[:, 0], x[:, 1],s=1, c='k')
plt.show()
```
